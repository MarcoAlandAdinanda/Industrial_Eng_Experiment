#### 2.4 Proses Analisis

```python
# Fungsi untuk menghitung percepatan
def acceleration(distance, time):
  return (2 * distance) / time**2

# Fungsi untuk menghitung koefisien gaya gesek
def kinetic_friction_coefficient(gravity, angle_degrees, acceleration):
  return (gravity * math.sin(math.radians(angle_degrees)) - acceleration) / (gravity * math.cos(math.radians(angle_degrees)))

# Fungsi untuk menghitung gaya gesek yang dihasilkan
def friction_force(friction_coef, mass, gravity, angle_degrees):
  return friction_coef * (mass * gravity) * math.cos(math.radians(angle_degrees))

# Fungsi untuk menghitung gaya kinetik yang dihasilkan
def kinetic_force(mass, gravity, angle_degrees):
  return (mass * gravity) * math.sin(math.radians(angle_degrees))
```

```python
# 1. Gunakan fungsi acceleration untuk menghitung percepatan tiap percobaan
df["Akselerasi (m/s^2)"] = df.apply(lambda row: acceleration(row['Panjang Lintasan (meter)'], row["Waktu (second)"]), axis=1)

# 2. Hitung koefisien gesek menggunakan fungsi fungsi kinetic_friction_coefficient
df["Koefisien Gesek"] = df.apply(lambda row: kinetic_friction_coefficient(9.8, row["Sudut Kemiringan (rad)"], row["Akselerasi (m/s^2)"]), axis=1)

# 3. Tentukan gaya yang dihasilkan oleh percepatan balok
df["Gaya (N)"] = df.apply(lambda row: kinetic_force(row["Massa (kilogram)"], 9.8, row["Sudut Kemiringan (rad)"]), axis=1)

# 3. Tentukan gaya gesek yang ditimbulkan oleh bidang
df["Gaya Gesek (N)"] = df.apply(lambda row: friction_force(row["Koefisien Gesek"], row["Massa (kilogram)"], 9.8, row["Sudut Kemiringan (rad)"]), axis=1)

# 4. Tentukan resultan gaya
df["Resultan Gaya (N)"] = df["Gaya (N)"] - df["Gaya Gesek (N)"]

# display
df
```

### 2.5 Hasil Analisis

```python
sudut_15 = df[df["Sudut Kemiringan (rad)"] == 15]
sudut_30 = df[df["Sudut Kemiringan (rad)"] == 30]
sudut_45 = df[df["Sudut Kemiringan (rad)"] == 45]
```

```python
### HASIL PENELITIAN
print("====================== HASIL ======================")
print(f"Rata - rata Koefisien Gesek          : {df['Koefisien Gesek'].mean()}")
print(f"Rata - rata Koefisien Gesek sudut 15 : {sudut_15['Koefisien Gesek'].mean()}")
print(f"Rata - rata Koefisien Gesek sudut 30 : {sudut_30['Koefisien Gesek'].mean()}")
print(f"Rata - rata Koefisien Gesek sudut 45 : {sudut_45['Koefisien Gesek'].mean()}")
print(f"Rata - rata Gaya Dorong     : {df['Gaya (N)'].mean()}")
print(f"Rata - rata Gaya Gesek      : {df['Gaya Gesek (N)'].mean()}")
print(f"Rata - rata Resultan Gaya   : {df['Resultan Gaya (N)'].mean()}")
```
